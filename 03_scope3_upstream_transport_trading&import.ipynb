```python
import pandas as pd

# Bundle trading DFs and emission factors
trading_dfs = {
    "cocoa": (trading_cocoa_df, cocoa_emission_factors),
    "coffee": (trading_coffee_df, coffee_emission_factors),
    "sesame": (trading_sesame_df, sesame_emission_factors),
    "nuts": (trading_nut_df, nut_emission_factors),
    "spice": (trading_spice_df, spice_emission_factors)
}

# Store emissions per product
product_emissions = []

# Loop and group only by Product Name (drop Category label)
for category, (df, ef_dict) in trading_dfs.items():
    emission_col = f"Emissions: {category.title()} (tCO2e)"
    if emission_col not in df.columns:
        continue

    # Group by product name
    grouped = df.groupby("Product Name")[emission_col].sum().reset_index()
    product_emissions.append(grouped)

# Combine all product-level results
scope3_1_product_df = pd.concat(product_emissions, ignore_index=True)

# Rename column to be consistent
emission_col_name = [col for col in scope3_1_product_df.columns if "Emissions" in col][0]
scope3_1_product_df.rename(columns={emission_col_name: "Total Emissions (tCO2e)"}, inplace=True)

# Sort and display
scope3_1_product_df.sort_values("Total Emissions (tCO2e)", ascending=False)
```
